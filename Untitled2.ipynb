## Data cleaning

### Subtask:
Clean the data loaded in the previous step.


**Reasoning**:
The data loading process has been unsuccessful in previous attempts.  I will try to clean the data using the last successfully loaded dataframe `df`.  I will handle missing values, outliers, and data type conversions.  I will also convert the 'Month of Period End' column to datetime objects.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Handling Missing Values
for col in df.columns:
    if df[col].isnull().sum() / len(df) > 0.5:
        print(f"Dropping column '{col}' due to more than 50% missing values.")
        df = df.drop(col, axis=1)
    elif df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# Outlier Treatment for 'Median Sale Price' and 'Homes Sold'
def treat_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    column = np.where(column < lower_bound, lower_bound, column)
    column = np.where(column > upper_bound, upper_bound, column)
    return column


# Convert 'Month of Period End' to datetime
df['Month of Period End'] = pd.to_datetime(df['Month of Period End'], errors='coerce')
df = df.dropna(subset=['Month of Period End'])


# Convert numerical columns to numeric types
for col in ['Median Sale Price', 'Homes Sold', 'Inventory', 'New Listings', 'Days on Market']:
    df[col] = df[col].astype(str).str.replace(r'[$,K]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = treat_outliers(df[col])

# Display info and head of the cleaned dataframe
df.info()
display(df.head())
```
